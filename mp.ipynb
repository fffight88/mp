```python
import folium
import pandas as pd
```

```python
import json
geo_path = 'seoul_gu.json'
geo_str = json.load(open(geo_path, encoding='utf-8'))
```

```python
geo_str
```

```python
pts = pd.read_csv("pts_gu.csv", encoding='cp949')
pts.head()
```

```python
ptsguindex = pts.set_index('구별')
ptsguindex.head()
```

```python
map = folium.Map(location=[37.5502, 126.982], zoom_start=11,)
fmap=folium.Choropleth(geo_data = geo_str,
               data = pts,
               columns = ['구별', '2012_평균'],
               fill_color = 'YlGnBu',
               key_on='feature.properties.name').add_to(map)

for s in fmap.geojson.data['features']:
    s['properties']['2012_average'] = ptsguindex.loc[s['properties']['name'], '2012_평균']
fmap.geojson.zoom_on_click = False
fmap.geojson.add_child(
    folium.features.GeoJsonTooltip(['name', '2012_average'],labels=False, 
                                   style=('background-color: black; color: white;font-weight:bold'))
)
display(map)
```
